```python
# matplotlib.use("ipympl")
res = torch.load(join(BEV_ROOT_DIR, "assets", "elevation_clip_-1_1_scale_0.05_prob_cluster.pt"))
HISTOGRAM_ELEVATION = torch.load(join(BEV_ROOT_DIR, "assets", "elevation_clip_-1_1_scale_0.05_prob_cluster.pt"))
HISTOGRAM_ELEVATION_EQ = torch.ones_like(HISTOGRAM_ELEVATION)
HISTOGRAM_RISK = torch.load(join(BEV_ROOT_DIR, "assets", f"wheel_risk_cvar_0_1_prob_cluster.pt"))
HISTOGRAM_RISK_EQ = torch.ones_like(HISTOGRAM_RISK)


tmp = 1 - (HISTOGRAM_ELEVATION / HISTOGRAM_ELEVATION.max())
WEIGHT_ELEVATION = tmp + HISTOGRAM_ELEVATION_EQ * 0.2
WEIGHT_ELEVATION /= WEIGHT_ELEVATION.max()

tmp = 1 - (HISTOGRAM_RISK / HISTOGRAM_RISK.max())
WEIGHT_RISK = tmp + HISTOGRAM_RISK_EQ * 0.2
WEIGHT_RISK /= WEIGHT_RISK.max()


def plot(name, y1, y2, sta=-25, sto=25 + 50 / 255, step=50 / 255):

    fig, ax = plt.subplots()
    labels = [f"{round(x,1)}" for x in np.arange(sta, sto, step)]
    ax.bar(labels, y1, align="edge", width=1.0, color=paper_colors_rgba_f["orange"], label="Probability")
    ax.set_ylabel("Probability")
    ax.set_title(name)

    ax.set_xticklabels
    ticks = ax.get_xticks()
    ax.set_xticks(ticks[::51])
    ticks = ax.get_yticks()
    ax.set_yticks(ticks[::2])

    x2 = np.arange(0, y1.shape[0])
    ax2 = ax.twinx()
    ax2.set_yticks(ax.get_yticks())
    ax2.set_yticklabels(ax.get_yticklabels())
    ax2.plot(x2, y2, color=paper_colors_rgba_f["magenta"], label="Weight")
    ax2.legend(title="Weight")
    ax2.set_ylabel("Weight")

    ax2.legend(loc="upper right", bbox_to_anchor=(0.5, 0.0, 0.455, 1.02))
    ax.legend(loc="upper right", bbox_to_anchor=(0.5, 0.0, 0.5, 0.96))
    # Show the plot

    ax2.set_yticks(np.linspace(ax2.get_yticks()[0], ax2.get_yticks()[-1], len(ax.get_yticks())))
    # ax.set_yticks(np.linspace(ax.get_yticks()[0], ax.get_yticks()[-1], len(ax2.get_yticks())))
    from perception_bev_learning import BEV_ROOT_DIR
    import os

    plt.savefig(os.path.join(BEV_ROOT_DIR, "assets", "balance_wheel_risk_cvar.svg"))
    plt.show()


plot("wheel_risk_cvar", HISTOGRAM_RISK, WEIGHT_RISK, sta=-1, sto=-1 + 1 / 100, step=1 / 100)
plot("elevation", HISTOGRAM_ELEVATION, WEIGHT_ELEVATION)
```
