```python
file_path_merged = "/home/patelm/Data/nature_hiking/2024-06-25-seealpsee/h5_1_reind_gt_2/run1.h5py"
```

```python
def get_time(hdf5_view, j=None):
    if j is None:
        return (
            np.array(hdf5_view["header_stamp_secs"])
            + np.array(hdf5_view["header_stamp_nsecs"]) * 10**-9
        )
    else:
        return (
            np.array(hdf5_view["header_stamp_secs"][j])
            + np.array(hdf5_view["header_stamp_nsecs"][j]) * 10**-9
        )


with h5py.File(file_path_merged, 'r') as f:
    # List all groups
    print("Keys: %s" % f.keys())
    keys = list(f.keys())
    dataset = f[keys[0]]
    travmap_key = "traversability_map_short_gt"
    print(dataset.keys())

    sample_idx = 288
    
    print(dataset[travmap_key].keys())
    print(get_time(dataset[travmap_key], j=sample_idx))  # This gives the timestamp 
    print(dataset[travmap_key]["position"][sample_idx]) # This is np array of size 3x1 (Same as the translation part of your HDF5 pose)
    print(dataset[travmap_key]["data"][sample_idx].shape) # This is the np array of shape (Layers , H, W)

```
